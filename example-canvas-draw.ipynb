```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
x = np.linspace(0, 10, 100)
```

```python
y1 = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1);
```

```python
fig.canvas.draw()
```

```python
plt.close(fig)
```

```python
y2 = np.cos(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2);
```

```python
fig.canvas.draw()
```

```python
plt.close(fig)
```
